```python
import pandas as pd
import time
pd.set_option('display.max_rows', None)
```

```python
standings_2025 = pd.read_html('https://www.mlb.com/standings/mlb?date=2025-07-15')
standings_2025 = pd.concat(standings_2025)
standings_2025 = standings_2025.rename(columns={'American League':'TEAM'})
standings_2025 = standings_2025[standings_2025.iloc[:, 0] != 'National League']
standings_2025["TEAM"][25] = "Athletics Athletics"


stats_2025 = pd.read_html("https://www.espn.com/mlb/stats/rpi/_/year/2025/sort/sos")
stats_2025 = pd.concat(stats_2025)
stats_2025 = stats_2025[2:]
stats_2025 = stats_2025.rename(columns={1:'TEAM'})


new_standings_2025 = pd.merge(standings_2025, stats_2025, on = 'TEAM', how = 'inner')
new_standings_2025 = new_standings_2025.rename(columns={2:'RPI', 6:'SOS',
                                                        10: 'EXPW-L', 11:'EXPWP'}).drop(columns=[0,3,4,5,7,8,9,12,'Next Game', 'STRK','L10'])
new_standings_2025['Home_Win_PCT'] = round((new_standings_2025['HOME'].str.split('-').str[0].astype(int) /(new_standings_2025['HOME'].str.split('-').str[0].astype(int) +
                                                                                                    new_standings_2025['HOME'].str.split('-').str[1].astype(int))), 3)

new_standings_2025['Away_Win_PCT'] = round((new_standings_2025['AWAY'].str.split('-').str[0].astype(int) /(new_standings_2025['AWAY'].str.split('-').str[0].astype(int) +
                                                                                                    new_standings_2025['AWAY'].str.split('-').str[1].astype(int))), 3)


```

```python
x =[]
all_star_dates = [
    "2024-07-17",
    "2023-07-13",
    "2022-07-19",
    "2021-07-14",
    '2020-09-13',
    "2019-07-08",
    "2018-07-16",
    "2017-07-10",
    "2016-07-11",
    "2015-07-13",
    "2014-07-14",
    "2013-07-17",
    "2012-07-09",
    "2011-07-11",
    "2010-07-12",
    "2009-07-13",
    "2008-07-14",
    "2007-07-09",
    "2006-07-10"
]

for i in range(1, 20):
  year = 2025 - i
  try:
    # Scrape standings from MLB
    standings_x = pd.read_html(f'https://www.mlb.com/standings/league/{year}?date={all_star_dates[i-1]}')
    standings_x = pd.concat(standings_x)
    standings_x = standings_x.rename(columns={'American League':'TEAM'})
    standings_x = standings_x[standings_x.iloc[:, 0] != 'National League']
    standings_x['Year'] = year
    standings_x['TEAM'] = standings_x['TEAM'].str.rstrip('zwyZWY')

    # Sleep before next request
    time.sleep(2)

    # Scrape RPI stats from ESPN
    stats_x = pd.read_html(f"https://www.espn.com/mlb/stats/rpi/_/year/{year}/sort/sos")
    stats_x = pd.concat(stats_x)
    stats_x = stats_x[2:]  # Skip header rows
    stats_x = stats_x.rename(columns={1: 'TEAM'})

    # Merge the two datasets
    new_standings_x = pd.merge(standings_x, stats_x, on='TEAM', how='inner')
    new_standings_x = new_standings_x.rename(columns={2:'RPI', 6:'SOS',
                                                      10: 'EXPW-L', 11:'EXPWP'}).drop(columns=[0,3,4,5,7,8,9,12, 'STRK','L10'])
    new_standings_x['Home_Win_PCT'] = round((new_standings_x['HOME'].str.split('-').str[0].astype(int) /(new_standings_x['HOME'].str.split('-').str[0].astype(int) +
                                                                                                    new_standings_x['HOME'].str.split('-').str[1].astype(int))), 3)
    new_standings_x['Away_Win_PCT'] = round((new_standings_x['AWAY'].str.split('-').str[0].astype(int) /(new_standings_x['AWAY'].str.split('-').str[0].astype(int) +
                                                                                                    new_standings_x['AWAY'].str.split('-').str[1].astype(int))), 3)
    x.append(new_standings_x)

    # Sleep again before the next year
    time.sleep(3)

  except Exception as e:
    print(f"Failed for year {year}: {e}")
    continue
```
